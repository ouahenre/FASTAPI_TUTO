```python
#import des librairies
#1 biologiste dit qu'Il ya  150 fleurs différentes classées selon 3 espèces  différentes  iris vergicolor, iris verginica, iris cetosa
#prédire une espèce connaissant ces caracteristiques : 
# Longeur et largeur : cépale et pétale

from sklearn.datasets import load_iris 
#Algorithme ML : tâche de classification
from sklearn.linear_model import LogisticRegression
#division jeu de données en 2 sous-ensembles: Test et entrainement
from sklearn.model_selection import train_test_split
import numpy as np
#fonction permettant enregistrer le modèle
from joblib import dump



```

```python
#charger les données : se présentant sous forme de dictionnaire: des variables indépendantes 
# (sepal width(cm),petal length(cm) petal width(cm) pour
#la variable cible
iris = load_iris()
iris
```

```python
#récupération des features: variables indépendantes
#features
X=iris.data
#target
Y=iris.target
```

```python
print(X)
print(Y)
```

```python
#algorithme de classification multiclasse
#division en data d'entraînement (pour entraîner l'algorithme et former le modèle)
# et test(pour évaluer le modèle)
```

```python
#Train/Test SPlit
#définir graîne aléatoire pour s'assurer de la réproductibilité des résultats
seed=13
X_train,X_test,y_train,y_test=train_test_split(
    X, Y,
    #%des données de test:20%pour évaluer le modèle formé et 80% pour entraîner le modèle
    test_size=0.2,
    random_state=seed
)
print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
```

```python
#ci-dessus j'ai 120 observations pour entraîner l'algorithme et 20 pour le tester
#formation du modèle de regression logistique:instanciation
logreg=LogisticRegression(
    multi_class="auto",
    #nombre d'itération:500 au minimum
    max_iter=500
)
#entraîner le modèle
logreg.fit(X_train,y_train)
```

```python
#voir le score
logreg.score(X_train,y_train)
```

```python
#ci-dessus: score d'entrainement de 0.975
```

```python
#score sur les données de test
#prédictions
y_pred=logreg.predict(X_test)
y_pred
```

```python
#calcul score sur les données de test
from sklearn.metrics import accuracy_score
accuracy_score(
    y_test,y_pred
)
```

```python
#sauvegarde du modèle
dump(logreg,'logreg.joblib')
```
